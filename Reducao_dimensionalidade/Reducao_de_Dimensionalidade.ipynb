Usando cv2 para reduzir dimensonalidade de fotos

```python
# Importando bibliotecas
import cv2
from google.colab.patches import cv2_imshow
```

```python
# Carregando imagem
foto = cv2.imread('Ciências Biológicas_327_DEK_0181.jpg')
foto
```

```python
# Convertendo para escala de cinza
imagem_cinza = cv2.cvtColor(foto, cv2.COLOR_BGR2GRAY)
cv2_imshow(imagem_cinza)
```

```python
# Converntendo para escala binária
_, imagem_binaria = cv2.threshold(imagem_cinza, 127, 255, cv2.THRESH_BINARY)
imagem_binaria
```

```python
# Salvando imagens
cv2.imwrite('imagem_cinza.jpg', imagem_cinza)
cv2.imwrite('imagem_binária.jpg', imagem_binaria)
```

Reduzindo dimensionalidade sem o auxílio de bibliotecas

```python
# Importando biblioteca
from PIL import Image
```

```python
# Carregando a imagem e extraindo largura e altura
imagem = Image.open('Ciências Biológicas_327_DEK_0181.jpg')
largura, altura = imagem.size
print(largura, altura)
```

```python
# Iniciando converção para escala de cinza
imagem_cinza = Image.new('L', (largura, altura))

# Percorrer cada píxel da imagem
for x in range(largura):
    for y in range(altura):
        # Obter os valores de R, G e B do pixel
        r, g, b = imagem.getpixel((x, y))

        # Calcular a média para tons de cinza
        cinza = int(0.3 * r + 0.59 * g + 0.11 * b)

        # Definir o pixel na imagem em tons de cinza
        imagem_cinza.putpixel((x, y), cinza)
```
